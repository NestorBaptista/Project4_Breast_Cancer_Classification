```python
file_names = ["/content/mass_case_description_test_set.csv", "/content/mass_case_description_train_set.csv" ]
df = pd.concat([pd.read_csv(file) for file in file_names])
df.head()
```

```python
# Splitting the 'image file path' column based on '/' and adding the extracted string to a new 'path' column
df['folder_path'] = df['image file path'].str.split('/').str[2]

# Drop unnecessary columns
df = df.drop(['image file path', 'cropped image file path', 'ROI mask file path'], axis=1)
df.head()
```

```python
df.dtypes
```

```python
# List of columns to check for blanks
columns_to_check = ['patient_id', 'breast_density', 'left or right breast',
                    'image view', 'abnormality id', 'abnormality type',
                    'mass shape', 'mass margins', 'assessment',
                    'pathology', 'subtlety', 'folder_path']

# Check for blanks in specified columns
blanks = df[columns_to_check].applymap(lambda x: isinstance(x, (str, bytes)) and x.isspace())

# Check if any blanks exist in the specified columns
if blanks.any().any():
    print("There are blanks in the specified columns.")
else:
    print("There are no blanks in the specified columns.")
```
